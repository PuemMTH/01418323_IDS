```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dic = {
    "day": ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"],
    "temp": [24,25,29,35,23,20,23]
}
df = pd.DataFrame(dic)
x = df["day"]
y = df["temp"]
```

```python
# ms = marker size
# mec = marker edge color
# mfc = marker face color
# ls = line style
# lw = line width
# c = color
plt.figure(figsize=(4, 3))
plt.plot(x, y, marker="*", ms=10, mec="r", mfc="y", ls="--", lw=2, c="#FFF000", label="Temp")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize=(4, 3))
# "maker,linemarker,linestyle"
plt.plot(x, y, "s-.b", label="Temp")
plt.legend()
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.plot(x, y, "s-.b", label="Temp")
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 12
plt.rcParams["font.weight"] = "bold"
plt.rcdefaults()
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.plot(x, y, "s-.b", label="Temp")
x = np.arange(0, 10, 0.01)-1
y = np.sin(x)
plt.plot(x, y, "s-.b", label="Temp")
plt.xticks(np.arange(0, 10, 1))
plt.yticks(np.arange(-1, 1, 0.2))
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.savefig("plot.png")
plt.show()
```

MultiGraph

```python
dic = {
    "day": ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"],
    "temp": np.random.randint(20, 35, 7),
    "temp2": np.random.randint(20, 35, 7)
}
df = pd.DataFrame(dic)
x = df["day"]
y = df["temp"]
y2 = df["temp2"]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.suptitle("Temperature", fontsize=16, fontweight="bold")

axes[0, 0].plot(x, y, "s-.b")
axes[0, 0].tick_params(axis="x", rotation=45)
axes[0, 0].set_title("Temp")

axes[0, 1].plot(x, y2, "s-.r")
axes[0, 1].tick_params(axis="x", rotation=45)
axes[0, 1].set_title("Temp")

axes[1, 0].plot(x, y, "s-.g")
axes[1, 0].tick_params(axis="x", rotation=45)
axes[1, 0].set_title("Temp")

axes[1, 1].plot(x, y2, "s-.y")
axes[1, 1].tick_params(axis="x", rotation=45)
axes[1, 1].set_title("Temp")

fig.tight_layout() # Adjust the spacing between subplots
fig.savefig("plot.png")
plt.show()

# Trick shift + tab to see the documentation
```
