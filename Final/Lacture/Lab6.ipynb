```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data/k-means.csv')
```

```python
df.head()
```

```python
# n_clusters คือ จำนวนกลุ่มที่ต้องการจะแบ่ง
# random_state คือ การกำหนดค่า seed ให้กับการสุ่ม
model = KMeans(n_clusters=3,random_state=0)
```

```python
# สร้าง model โดยใช้ข้อมูลจาก df
model.fit(df)
```

```python
# ดูจุดศูนย์กลางของแต่ละกลุ่ม
model.cluster_centers_
```

```python
# figure size กว้าง 4 นิ้ว สูง 3 นิ้ว
plt.figure(figsize=[4,3])
```

```python
# scatter plot ของข้อมูล และจุดศูนย์กลางของแต่ละกลุ่ม
plt.scatter(df['x'], df['y'])
```

```python
# ดึงจุดศูนย์กลางของแต่ละกลุ่มมาเก็บไว้ในตัวแปร CT
CT = model.cluster_centers_

# scatter plot ของจุดศูนย์กลางของแต่ละกลุ่ม
plt.scatter(CT[:,0],CT[:,1], c='black')
```

```python
tc = ['red', 'blue', 'green']
point_color = []
# นำ label ของแต่ละจุดมาเก็บไว้ใน point_color
for i in model.labels_ :
    point_color.append(tc[i])
```

```python
plt.figure(figsize=[4,3])

# scatter plot ของข้อมูล และจุดศูนย์กลางของแต่ละกลุ่ม
plt.scatter(df['x'], df['y'], c = point_color)
```

```python
plt.figure(figsize=[4,3])
plt.scatter(df['x'], df['y'], c=point_color, marker = '*', s = 150)
```

```python
# Code นี้ทำการ import library และดึงข้อมูลจากไฟล์ 
# k-means.csv และใช้ algorithm KMeans ที่มีจำนวน cluster เท่ากับ 3 และ random state เท่ากับ 0 
# ในการทำ clustering จากนั้นทำการ fit ข้อมูลด้วย model
# และจากนั้นทำการพล็อต scatter plot ของข้อมูล 
# โดยให้แต่ละจุดมีสีที่แตกต่างกันตาม cluster ที่ model 
# ทำการจัดกลุ่มไว้ และกำหนดสีของแต่ละ cluster ด้วย tc ซึ่งเป็น list 
# ที่เก็บสีแต่ละ cluster
# สุดท้ายทำการ plot จุดที่เป็นจุดศูนย์กลางของแต่ละสี

from sklearn.cluster import KMeans
df = pd.read_csv('data/k-means.csv')
model = KMeans(n_clusters=3, random_state = 0)
model.fit(df)
tc = ['red', 'blue', 'green']
point_color = []

for i in model.labels_ :
    point_color.append(tc[i])
plt.figure(figsize=[4,3])
plt.scatter(df['x'],df['y'], c = point_color)

CT = model.cluster_centers_
plt.scatter(CT[:,0], CT[:,1], c='black', marker = '*', s = 150)
```

```python
# สร้างข้อมูลที่มีจุดที่อยู่ใน cluster ที่ 0 และ 1 และ 2
test_df = pd.DataFrame([[50,50]] ,columns = ['x','y'])
# ทำการ predict จุดที่ต้องการทราบว่าอยู่ใน cluster ไหน
model.predict(test_df)
```

```python
plt.figure(figsize=[4,3])
# ทำการ plot ข้อมูลทั้งหมด
plt.scatter(df['x'], df['y'], c='black', marker = '*', s = 150)
plt.scatter(test_df['x'],test_df['y'], c='red')
```

```python
from sklearn.preprocessing import StandardScaler
# สร้าง scaler เพื่อทำการ scale ข้อมูล
scaler = StandardScaler()
# ทำการ scale ข้อมูลใน df โดยใช้ scaler ที่สร้างไว้
df['x'] = scaler.fit_transform(df[['x']])
df['y'] = scaler.fit_transform(df[['y']])
```

```python
df = pd.read_csv('data/k-means.csv')
# สร้าง scaler เพื่อทำการ scale ข้อมูล
scaler = StandardScaler()
df['x'] = scaler.fit_transform(df[['x']])
df['y'] = scaler.fit_transform(df[['y']])

# สร้าง model โดยใช้ข้อมูลจาก df
model = KMeans(n_clusters=3, random_state = 0)
# ทำการ fit ข้อมูลด้วย model
model.fit(df)
tc = ['red', 'blue', 'green']
point_color = []

# นำ label ของแต่ละจุดมาเก็บไว้ใน point_color
for i in model.labels_ :
    point_color.append(tc[i])
plt.figure(figsize=[4,3])
plt.scatter(df['x'],df['y'], c = point_color)

# ดึงจุดศูนย์กลางของแต่ละกลุ่มมาเก็บไว้ในตัวแปร CT
CT = model.cluster_centers_
plt.scatter(CT[:,0], CT[:,1], c='black', marker = '*', s = 150)
```
