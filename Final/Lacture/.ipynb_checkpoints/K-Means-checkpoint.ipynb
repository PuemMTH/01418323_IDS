```python
df = pd.read_csv('./data/test/bank-small.csv', sep=';')
df = df.sample(n=2000)
model = KMeans(n_clusters=3,random_state=0)
model_df = df[['age','balance']]

model.fit(model_df[['age','balance']])
model.cluster_centers_

plt.figure(figsize=(5,5))
plt.scatter(model_df['age'],model_df['balance'],c=model.labels_)
plt.show()
```
